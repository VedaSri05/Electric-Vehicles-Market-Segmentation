### Libraries Used

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step-1: Load and Explore the Dataset

```python
# Load the dataset
data = pd.read_csv('dataset/EV_Dataset.csv')

# Display the first few rows of the dataset to get an overview of the data
data.head()
```

### Step-2: Check for Missing Values

```python
# Check for missing values in the dataset
print(data.isnull().sum())
```

#### There are no missing values in the dataset, so the handling of missing values is not necessary

### Step-3: Filter Data

```python
# Check unique values in the 'Vehicle_Category' column
print(data['Vehicle_Category'].unique())
```

```python
# Aggregate EV sales quantity by Vehicle Category
vehicle_category_sales = data.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum().reset_index()

# Sort the values by sales quantity in descending order
vehicle_category_sales_sorted = vehicle_category_sales.sort_values(by='EV_Sales_Quantity', ascending=False)

# Display the result
vehicle_category_sales_sorted
```

### Step-4: Visualize the Aggregated Data

```python
# Plot the total sales for each Vehicle Category
plt.figure(figsize=(5, 5))
plt.bar(vehicle_category_sales_sorted['Vehicle_Category'], vehicle_category_sales_sorted['EV_Sales_Quantity'], color='green')
plt.title('Total EV Sales Quantity by Vehicle Category')
plt.xlabel('Vehicle Category')
plt.ylabel('Total EV Sales Quantity')
plt.xticks(rotation=45)
plt.show()
```

### Filter for 2-Wheelers in Vehicle Category

```python
# Filter the data to focus only on 2-Wheelers
two_wheelers_data = data[data['Vehicle_Category'] == '2-Wheelers']
```

### Aggregate by Year

```python
# Aggregate sales by Year for 2-Wheelers
two_wheelers_sales_year = two_wheelers_data.groupby('Year')['EV_Sales_Quantity'].sum().reset_index()

# Sort sales by year in descending order
two_wheelers_sales_year_sorted = two_wheelers_sales_year.sort_values(by='EV_Sales_Quantity', ascending=False)

# Display the result
two_wheelers_sales_year_sorted
```

### Aggregate by Month

```python
# Aggregate sales by Month_Name for 2-Wheelers
two_wheelers_sales_month = two_wheelers_data.groupby('Month_Name')['EV_Sales_Quantity'].sum().reset_index()

# Sort sales by month in descending order
two_wheelers_sales_month_sorted = two_wheelers_sales_month.sort_values(by='EV_Sales_Quantity', ascending=False)

# Display the result
two_wheelers_sales_month_sorted
```

### Plotting Sales by Year

```python
# Plot the total sales of 2-Wheelers by Year
plt.figure(figsize=(5, 5))
plt.bar(two_wheelers_sales_year_sorted['Year'], two_wheelers_sales_year_sorted['EV_Sales_Quantity'], color='blue')
plt.title('Total EV Sales Quantity for 2-Wheelers by Year')
plt.xlabel('Year')
plt.ylabel('Total EV Sales Quantity')
plt.xticks(rotation=45)
plt.show()
```

### Plotting Sales by Month

```python
# Plot the total sales of 2-Wheelers by Month
plt.figure(figsize=(8, 6))
plt.bar(two_wheelers_sales_month_sorted['Month_Name'], two_wheelers_sales_month_sorted['EV_Sales_Quantity'], color='green')
plt.title('Total EV Sales Quantity for 2-Wheelers by Month')
plt.xlabel('Month')
plt.ylabel('Total EV Sales Quantity')
plt.xticks(rotation=45)
plt.show()
```

### Visualization of Market Share Across States

```python
# Assuming EV dataset has 'State' and 'EV_Sales_Quantity' columns
state_sales = data.groupby('State')['EV_Sales_Quantity'].sum()
```

```python
# Plotting
plt.figure(figsize=(12, 8))
state_sales.sort_values(ascending=False).plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Total EV Sales by State')
plt.xlabel('State')
plt.ylabel('Total EV Sales Quantity')
plt.xticks(rotation=90)
plt.show()
```

# Electric Vehicle Sales Analysis

## Overview:
In this analysis, we focused on the sales of **2-Wheelers** electric vehicles (EVs) from the dataset. The goal is determined which is a **2-Wheelers** category (e.g., electric bikes, scooters) had the highest sales in recent years and months.

## Key Findings:
- After aggregating and analyzing the data, we found that **2-Wheelers** consistently led the EV sales in terms of quantity.
- Upon further investigation into sales trends, we observed that **2023** witnessed a peak in **2-Wheeler** sales, with the highest number of units sold in **November 2023**.

### Summary of the Sales in 2023:
- The year **2023** marked a significant growth in electric 2-wheeler sales, indicating an increasing demand for electric bikes and scooters in India.
- **November** stands out as the month with the highest sales, showcasing a potential spike in consumer interest, possibly due to promotions, festivals, or other market factors.

## Conclusion:
Based on the data analysis, **2-Wheelers** remain the dominant category in the electric vehicle market, with the most significant sales observed in **November 2023**. This highlights the growing acceptance and adoption of electric vehicles, especially 2-wheelers, among Indian consumers.
